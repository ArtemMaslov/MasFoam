### Тестирование методов Эйлера

Задача: равноускоренное движение материальной точки (диф. ур. 1 порядка для скорости).
Граничные условия: выбираются случайным образом.
Строятся графики: график ошибки от времени {log(res)}(t) и координаты от времени x(t).

Выводы:
1. С увеличением времени ошибка увеличивается по порядку от 1e-8 до 1e2.

```python
v0 = 1 # m/s
w = 1 # m/s

def funct1(t, x):
    global v0
    global w
    return v0 + w * t + 330 * t**2 + 141 * t**3

def funct1_prec(t, x0, t0):
    global v0
    global w
    delta = v0 * t0 + w / 2 * t0**2 + 330 * t0**3 / 3 + 141 * t0**4 / 4
    return x0 + v0 * t + w / 2 * t**2 + 330 * t**3 / 3 + 141 * t**4 / 4 - delta

def TestEuler1():
    global x0
    global v0
    global w

    points_count = 1
    plots_count = 3

    # fig, axes = plt.subplots(plots_count)
    # fig.set_size_inches(10, 24)
    # fig.suptitle("Тестирование явного метода Эйлера. Равноускоренное движение.")
    # axes[len(axes) - 1].set(xlabel = "t")

    # fig1, axes1 = plt.subplots(plots_count)
    # fig1.set_size_inches(10, 24)
    # fig1.suptitle("Тестирование явного метода Эйлера. Равноускоренное движение.")
    # axes1[len(axes1) - 1].set(xlabel = "t")

    for st in range(0, plots_count):
        # граничные условия
        x0 = random.random() + random.randint(-1e3, 1e3)
        w = random.random() + random.randint(-1e3, 1e3)
        v0 = random.random() + random.randint(-1e3, 1e3)
        t1 = random.random() + random.randint(-1e3, 1e3)
        dt = random.random() + random.randint(0, 1e3)
        #t2 = t1 + dt
        t2 = t1 + 1e-2

        # сетка
        tau = (t2 - t1) / points_count
        t = numpy.arange(t1, t2 + tau, tau)
        
        x = DifEqSolvers.DifEqEulerSolver2a(funct1, t, x0)
        xb = DifEqSolvers.DifEqEulerSolver2b(funct1, t, x0)
        #x2 = DifEqSolvers.DifEqEulerSolver2a(funct1, t, x0)
        x_prec = funct1_prec(t, x0, t[0])

        print(f"x0 = {x0}")
        print(f"x[0] = {x[0]}")
        print(f"x_prec[0] = {x_prec[0]}")
        print(f"max_res = {math.log10(max(numpy.abs(x_prec - x)))}")
        print(f"max_resb = {math.log10(max(numpy.abs(x_prec - xb)))}")
        print("")
        break

        # axes[st].plot(t, numpy.log10(numpy.abs(x_prec - x)), label = "Euler 1a")
        # axes[st].plot(t, numpy.log10(numpy.abs(x_prec - xb)), label = "Euler 1b")
        # axes[st].set_title(f"$v_0 = {v0}$, $w = {w}$, $x_0 = {x0}$")
        # axes[st].set(ylabel = "$\log{res}$")
        # axes[st].legend()

        # axes1[st].plot(t, x, label = "Euler 1a")
        # axes1[st].plot(t, xb, label = "Euler 1b")
        # axes1[st].plot(t, x_prec, label = "Precision")
        # axes1[st].set_title(f"$v_0 = {v0}$, $w = {w}$, $x_0 = {x0}$")
        # axes1[st].set(ylabel = "$x$")
        # axes1[st].legend()

TestEuler1()
```
